sdg-score-multi__number,sdg-score-multi__number 2,sdg-score-multi__number 3,sdg-score-multi__number 5,sdg-score-multi__number 6,sdg-score-multi__number 7,sdg-score-multi__number 8

we build checkerr method to check that

```python
#as dg-score-multi__number represent the sdg gaol number we want to make sure the same goal is not duplicated for the same row

def checkerr(roww):
  lst=[i for i in roww.tolist() if pd.notna(i)]
  sett=set(lst)
  for i in sett:
    if lst.count(i) > 1:
      return False
  return True


```

clean  sdg-score-multi__number,sdg-score-multi__value,
sdg-score-multi__number 2,sdg-score-multi__value 2,
sdg-score-multi__number 3,sdg-score-multi__value 3,sdg-score-multi__value 4 columns:

we have observed sdg_number contains the goal number
sdg_value contains the value of that goal since most of the 8 right columns are null we have to move cells that are not null in the rightmost such that the goal number and related value nonnull into leftmost columns that contain null cells in the same sdg_number and sdg_value

```python
# move first nonnull number and value of right most columns to sdg-score-multi__number,sdg-score-multi__value that they are null
def update_scores(row):
  # the number of right most columns from sdg 5 to sdg8 of number and value
    for i in range(5, 9):
        number_col = f'sdg-score-multi__number {i}'
        value_col = f'sdg-score-multi__value {i}'

        if pd.notna(row[number_col]) and pd.notna(row[value_col]):
            # Update the columns
            row['sdg-score-multi__number'] = row[number_col]
            row['sdg-score-multi__value'] = row[value_col]

            # Set the number and value to NaN in right most column
            row[number_col] = np.nan
            row[value_col] = np.nan

            # Break to prevent checking subsequent columns for this row
            break

    return row
```

```python
THE.loc[THE['sdg-score-multi__number'].isnull() & THE['sdg-score-multi__value'].isnull()]=THE[THE['sdg-score-multi__number'].isnull() & THE['sdg-score-multi__value'].isnull()].apply(update_scores,axis=1)
```

```python
# move first nonnull number and value of right most columns to sdg-score-multi__number2,sdg-score-multi__value2 that they are null

def update_scores2(row):
    for i in range(5, 9):
        number_col = f'sdg-score-multi__number {i}'
        value_col = f'sdg-score-multi__value {i}'

        if pd.notna(row[number_col]) and pd.notna(row[value_col]):
            # Update the columns
            row['sdg-score-multi__number 2'] = row[number_col]
            row['sdg-score-multi__value 2'] = row[value_col]

            # Set number and value columns to NaN
            row[number_col] = np.nan
            row[value_col] = np.nan

            # Break to prevent checking subsequent columns for this row
            break

    return row
```

```python
THE.loc[THE['sdg-score-multi__number 2'].isnull() & THE['sdg-score-multi__value 2'].isnull()] = THE[THE['sdg-score-multi__number 2'].isnull() & THE['sdg-score-multi__value 2'].isnull()].apply(update_scores2,axis=1)
```

```python
# move first nonnull number and value of right most columns to sdg-score-multi__number3,sdg-score-multi__value3 that they are null

def update_scores3(row):
    for i in range(5, 9):
        number_col = f'sdg-score-multi__number {i}'
        value_col = f'sdg-score-multi__value {i}'

        if pd.notna(row[number_col]) and pd.notna(row[value_col]):
            # Update the columns
            row['sdg-score-multi__number 3'] = row[number_col]
            row['sdg-score-multi__value 3'] = row[value_col]

            # Set the number and value columns to NaN
            row[number_col] = np.nan
            row[value_col] = np.nan

            # Break to prevent checking subsequent columns for this row
            break

    return row
```
